Exploratory Data Analysis (EDA) and Visualization
Campaign Performance:

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_csv,x='Acquisition_Cost',y='ROI')
plt.title('Relationship between Acquisition_Cost and ROI')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Channel_Used',y='Conversion_Rate',hue='Campaign_Type',data=df_csv)
plt.title('Average Conversion_Rate for different Channel_Used by Campaign_Type')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=df_csv,x='Campaign_Type',y='Engagement_Score')
plt.title('Distribution of Engagement_Score across Campaign_Type')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_csv,x='Company',y='ROI',hue='Company')
plt.title('Average ROI by Company')
plt.show()
```

```python
corr_matrix=df_csv[['Engagement_Score','Conversion_Rate']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

Customer Segmentation:

```python
ax=sns.countplot(data=df_csv,x='Target_Audience',hue='Target_Audience')
ax.bar_label(ax.containers[0])
plt.title('Distribution of Target_Audience')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_csv,x='Customer_Segment',y='Conversion_Rate',hue='Language')
plt.title('highest Conversion_Rate for each Language')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=df_csv,x='Customer_Segment',y='Acquisition_Cost',hue='Channel_Used')
plt.title('Distribution of Acquisition_Cost across each Customer_Segment by Channel_Used')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_csv,x='Language',y='Conversion_Rate',hue='Language')
plt.title('Average Conversion_Rate by Language')
plt.show()
```

Channel Effectiveness

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_csv,x='Channel_Used',y='Engagement_Score',hue='Campaign_Type')
plt.title('Engagement_Score for different Channels_Used by Campaign_Type,')
plt.show()
```

```python
df_csv.groupby('Channel_Used')['ROI'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
sns.scatterplot(data=df_csv,x='Clicks',y='Impressions',hue='Channel_Used')
plt.title('Relationship between Clicks and Impressions')
plt.show()
```

Time-Based Analysis

```python
sns.histplot(df_csv['Duration'],kde=True,color='green')
plt.title('distribution of Duration')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=df_csv,x='Date',y='Conversion_Rate',hue='Company',marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=df_csv,x='Date',y='Engagement_Score',marker='o')
plt.xticks(rotation=45)
plt.show()
```

Geographic Analysis


```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_csv,x='Location',y='Acquisition_Cost',hue='Location')
plt.title('Highest Acquisition_Cost by location')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_csv,x='Location',y='Conversion_Rate',hue='Target_Audience')
plt.title('Conversion_Rate by different Location')
plt.show()
```

```python
df_csv.groupby('Location')['ROI'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```
